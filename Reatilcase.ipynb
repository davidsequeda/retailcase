tablas de frecuencia
para la soluci´´on del caso de suso es necesario identificas una posible hipótesis  para validar a través de las herramientas que se revisarán en las siguientes líneas

la tabla de frecuencia es una herramienta que permite ordenar los datos, de manera que se presentan numéricamente las caracteristicas de la distribución de un conjunto de datos  o muestra 

hipótesis: las categorías de celulares es la que más productos vende.
se queire validar ya que se desea lanzar una campaña de descuento para traer más lcientes a la empresa, sin embargo se debe conocer cuál va a ser la categoría en la cual se invierta más publicidad.


```python
#tabla de frecuencia univariada
#se va a realizar una tabla para analizar las categorías que
#se venden en el comerciop

frec_datos=datos["nombre_categoria_producto"].value_counts().reset_index()
frec_datos.head(8)
frec_datos.dtypes
```

```python
#frecuencia relativa
frec_datos["fre_relative_categoria_prodcuto"]=frec_datos["count"].apply(lambda x: x/frec_datos["count"].sum())
frec_datos.head()
```

```python
#frecuencia absoluta y relatica acumulada con sumsum()
#absoluta acumulada
frec_datos["frec__acumu_abs_categoria"]=frec_datos["count"].cumsum()
#relatica acumulada
frec_datos["frec_acum_rela_categoria"]=frec_datos["fre_relative_categoria_prodcuto"].cumsum()
frec_datos.head(20)
```

análisis:la tabla de frecuencias nos muestra las siguiente información

1.frecuencia absoluta: númeor de veces que aparece un valor
2.frecuencia relativa:es el resultaod de dividir la frecunecia absoluta de un producto por el número total de datos
frecuencia acumulada absoluta: la suma de las frecunecias absolutas
frecuencoas acumulada relativa: el resultado de dividir la frecuencia acumulad apor el n´+umero total de informacion

```python
#tabla de frecuencia bivariada
##tabla que muestra la frecuencia por ciudad y categoría del producto
pd.crosstab(index=datos["nombre_categoria_producto"],columns=datos["precio"])
```

## tendencia central
la tendencia central significa medir el centro de distribucion de ubicacion de valores de un conjunto de datos. Da una idea del valor promedio de las viraiables en el conjunto de datos y también una indicación de cuán ampliamente se distribuyen los valores en el conjunto de datos.
1. media
2. mediana
3. moda

Hipótesis : los precio en el comercio son mayores a 500, ya que el estudio de marketing arrojó que los clientes más recurrentes compran mayormente productos más económicos( debajo de 500), por lo tanto es necesario validar si en promedio la compañis tiene precios más competitivos

```python
# media o prmedio
#promedio de precio
datos["precio"].mean()
```

```python
#mediana
#mediana de precio, valor central de la ddistribución de los precio
datos["precio"].median()
```

```python
#moda
#mvalor más común en los precio
datos["precio"].mode()
```

Análisis: Es posible analizar que el promedio es inferior a 500. ubicandose en 120, la mediana que es el dato central se ubica en 83,12 y existen varios valores comunes y todo debajo de 60 lo que nos muestra que la mayor cantidad de precios son bajos y permiten ser más atractivos para los consumidores finales que compran mayormente precios debajo de 500

## medidas de localizacion
los cuantiles y percentiles son punetos de corte que dividen el rango de una distribución de probabilidad en intervalos continuos con probabilidades iguales o que dividen las observaciones en una muestra de la misma forma

Así pudes representar un porcentaje de datos a travpes de los cuantiles para identificar como estanlocalizado
